```python
## User Profile
```

```python
current_skills = ["Python","SQL"]

interest = "Artificial Intelligence"

goal = "Machine Learning Engineer"

difficulty = "Beginner"

study_time = 1

learning_style = "Video"
```

```python
## Career Skills Dictionary
```

```python
career_skills = {

"Machine Learning Engineer":
["Python","SQL","Statistics","Machine Learning","Deep Learning"],

"Data Scientist":
["Python","SQL","Statistics","Machine Learning","Power BI"],

"Data Analyst":
["Excel","SQL","Python","Power BI"]

}
```

```python
## Skill Gap Analysis
```

```python
required_skills = career_skills[goal]

missing_skills = list(set(required_skills)-set(current_skills))

print("Missing Skills")

print(missing_skills)
```

```python
## Course Recommendation
```

```python
print("Recommended Courses")

for skill in missing_skills:

    courses = df[
        df['Gained Skills'].str.contains(
            skill,
            case=False,
            na=False
        )
    ]

    print("\nSkill:", skill)

    if len(courses) > 0:
        print(courses['Title'].head(3).to_string(index=False))
    else:
        print("No course found")
```

```python
## Study Time Based Roadmap
```

```python
print("Learning Roadmap")

month = 1

for skill in missing_skills:

    print("Month",month,"→",skill)

    month += 1
```
